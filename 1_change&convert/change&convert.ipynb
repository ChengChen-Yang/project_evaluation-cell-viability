```python
import os 
import cv2
```

# 更改檔名(tif直接改附檔名會編碼錯誤)

```python
path = './img_input/'
save_path = './img_output'
ori_type = '.tif'
target_tpye = '.jpg'

files = os.listdir(path)
print(files)
n = 0
for i in files:
    oldname = path + files [n]
    newname = path +str(n+1)+ori_type
    os.rename(oldname, newname)
    #     print(oldname)
#     print(newname)
    n = n+1
```

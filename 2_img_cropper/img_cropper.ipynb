# 分隔圖片為4*3，以增加數據量

```python
path = './img_input/'
save_path = './img_output/' 


img_list = os.listdir(path)
# print(img_list)

n = 0
for a in img_list:
    files_path = path + img_list[n]
#     print(files_path)
    img = Image.open(files_path)
#     print(img)
    img_size = img.size
    print ( "圖片寬度和高度分別是{} " .format(img_size))
    xx = 4
    yy = 3
    x = img_size[0]//xx
    y = img_size[1]//yy
    for j in range (yy):
        for i in range (xx):
            left = i*x
            up = y*j
            right = left + x
            low = up +y
            region = img.crop((left,up,right,low))
            print((left,up,right,low))
            temp = str(i)+str(j)
            region.save(save_path+img_list[n].split('.')[0]+'_'+temp+'.jpg')
    n=n+1
```
